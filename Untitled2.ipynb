```python
import zipfile
import os

# Tentukan lokasi file zip Anda
zip_file_path = '/content/archive.zip'

# Tentukan folder tempat Anda ingin menyimpan hasil ekstraksi
extraction_path = '/content/dataset'

# Buat folder baru jika belum ada
if not os.path.exists(extraction_path):
    os.makedirs(extraction_path)

# Ekstrak semua file ke folder yang sudah dibuat
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"File berhasil diekstrak ke: {extraction_path}")
```

```python
# Tampilkan isi folder hasil ekstraksi
print("Isi folder:")
print(os.listdir(extraction_path))
```

```python
import pandas as pd
import os

# Tentukan folder tempat Anda menyimpan hasil ekstraksi
extraction_path = '/content/dataset'

# Ganti dengan nama file yang Anda pilih
file_name = 'shopping_trends_updated.csv'
full_path = os.path.join(extraction_path, file_name)

try:
    df = pd.read_csv(full_path)
    print("DataFrame berhasil dibuat!")
    print(df.head()) # Tampilkan 5 baris pertama untuk verifikasi
except FileNotFoundError:
    print(f"ERROR: File '{file_name}' tidak ditemukan di '{extraction_path}'.")
except Exception as e:
    print(f"Terjadi kesalahan lain: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur style visualisasi
sns.set_style("whitegrid")

# Buat bar chart dari frekuensi kategori
plt.figure(figsize=(10, 6))
df['Category'].value_counts().plot(kind='bar', color=sns.color_palette("viridis", len(df['Category'].unique())))
plt.title('Jumlah Produk per Kategori', fontsize=16)
plt.xlabel('Kategori Produk', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Buat histogram untuk kolom Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Usia Pembeli', fontsize=16)
plt.xlabel('Usia', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pastikan Anda telah mengganti dengan nama kolom yang benar
age_column = 'Age'
purchase_column = 'Purchase Amount (USD)' # <-- Ganti dengan nama kolom yang benar

# Langkah 1: Bersihkan kolom jumlah pembelian
# Hapus tanda dolar ($) dan ubah ke format numerik
df[purchase_column] = df[purchase_column].astype(str).str.replace('$', '').str.replace(',', '').str.strip()
df[purchase_column] = pd.to_numeric(df[purchase_column])

# Langkah 2: Buat scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=age_column, y=purchase_column, data=df, alpha=0.6, color='coral')
plt.title('Hubungan antara Usia dan Jumlah Pembelian', fontsize=16)
plt.xlabel('Usia', fontsize=12)
plt.ylabel('Jumlah Pembelian', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pastikan Anda telah membersihkan kolom Purchase_Amount sebelumnya
# df['Purchase Amount (USD)'] = pd.to_numeric(df['Purchase Amount (USD)'])

# Hitung rata-rata pengeluaran untuk setiap gender
average_purchase_by_gender = df.groupby('Gender')['Purchase Amount (USD)'].mean().reset_index()

# Buat bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Purchase Amount (USD)', data=average_purchase_by_gender, palette='viridis')
plt.title('Rata-Rata Pengeluaran Berdasarkan Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Rata-Rata Pengeluaran (USD)', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Hitung frekuensi setiap metode pembayaran
payment_counts = df['Payment Method'].value_counts()

# Buat pie chart
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribusi Metode Pembayaran', fontsize=16)
plt.axis('equal') # Memastikan pie chart berbentuk lingkaran
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung frekuensi setiap kategori pembelian
frequency_counts = df['Frequency of Purchases'].value_counts()

# Buat bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=frequency_counts.index, y=frequency_counts.values, palette='viridis')
plt.title('Frekuensi Pembelian Pelanggan', fontsize=16)
plt.xlabel('Frekuensi Pembelian', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
